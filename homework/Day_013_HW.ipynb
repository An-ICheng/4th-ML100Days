## 作業
1. 請將 app_train 中的 CNT_CHILDREN 依照下列規則分為四組，並將其結果在原本的 dataframe 命名為 CNT_CHILDREN_GROUP
    * 0 個小孩
    * 有 1 - 2 個小孩
    * 有 3 - 5 個小孩
    * 有超過 5 個小孩

2. 請根據 CNT_CHILDREN_GROUP 以及 TARGET，列出各組的平均 AMT_INCOME_TOTAL，並繪製 baxplot
3. 請根據 CNT_CHILDREN_GROUP 以及 TARGET，對 AMT_INCOME_TOTAL 計算 [Z 轉換](https://en.wikipedia.org/wiki/Standard_score) 後的分數

```python
#1
"""
Your code here
"""
cut_rule = [-np.inf, 0, 2, 5, np.inf] # 負無限大到0再到2再到5再到正無限大
# app_train['CNT_CHILDREN'].values # 是array，因為pd.cut，裡面要放array
app_train['CNT_CHILDREN_GROUP'] = pd.cut(app_train['CNT_CHILDREN'].values, cut_rule, include_lowest=True) # app_train['CNT_CHILDREN_GROUP'] 這樣就建了一個新的欄位
# app_train['CNT_CHILDREN_GROUP']
app_train['CNT_CHILDREN_GROUP'].value_counts() # printvalue_counts()是計算app_train['CNT_CHILDREN_GROUP']裡的每個值有幾個，如(2.0, 5.0]這個值有4230個
```

```python
#2-1
"""
# 根據 Day_013_dataFrame_operation 下面code也可寫成這樣
grp = ['CNT_CHILDREN_GROUP','TARGET']
app_train.groupby(grp)['AMT_INCOME_TOTAL'].mean()
# 或這樣
app_train.groupby(['CNT_CHILDREN_GROUP','TARGET'])['AMT_INCOME_TOTAL'].mean()
"""
grp = ['CNT_CHILDREN_GROUP','TARGET']
grouped_df = app_train.groupby(grp)['AMT_INCOME_TOTAL']
grouped_df.mean() # 當生0~2個小孩時，且TARGET是1也就是有活下來的人的平均收入AMT_INCOME_TOTAL為176115.362720
```

```python
#2-2
"""
Your code here
"""
plt_column = 'AMT_INCOME_TOTAL'
plt_by = ['CNT_CHILDREN_GROUP', 'TARGET']

app_train.boxplot(column=plt_column, by = plt_by, showfliers = False, figsize=(12,12)) # showfliers是否在圖中顯示離群值。reference:https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.boxplot.html
plt.suptitle('aaaaa')
plt.show()
```

```python
#3
"""
Your code here
"""
# ?In[5]grouped_df已經是group by後的dataframe，照理來說row會減少，為什麼執行後結果仍有307511個值
app_train['AMT_INCOME_TOTAL_Z_BY_CHILDREN_GRP-TARGET'] = grouped_df.apply(lambda x: (x-np.mean(x))/np.std(x))

app_train[['AMT_INCOME_TOTAL','AMT_INCOME_TOTAL_Z_BY_CHILDREN_GRP-TARGET']]
```
