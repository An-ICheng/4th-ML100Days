```python
# 做完特徵工程前的所有準備 (與前範例相同)
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LinearRegression

data_path = '/Users/anncheng/Documents/GitHub/4th-ML100Days/data/'
df_train = pd.read_csv(data_path + 'house_train.csv.gz')

train_Y = np.log1p(df_train['SalePrice'])
df = df_train.drop(['Id', 'SalePrice'] , axis=1)
df.head()
```

```python
#只取 int64, float64 兩種數值型欄位, 存於 num_features 中
num_features = []
for dtype, feature in zip(df.dtypes, df.columns):
    if dtype == 'float64' or dtype == 'int64':
        num_features.append(feature)
print(f'{len(num_features)} Numeric Features : {num_features}\n')
type(num_features)
```

```python
# 削減文字型欄位, 只剩數值型欄位
df = df[num_features] # df[columhn_name]印出columhn_name 這個欄位裡面的數值，而df[list]印出？？？
df = df.fillna(-1)
MMEncoder = MinMaxScaler()
# train_Y.shape[0] # 看第0行有幾個row
train_num = train_Y.shape[0]
df.head()
```

# 作業1
* 試著限制 '1樓地板面積(平方英尺)' (1stFlrSF) 欄位的上下限, 看看能否再進一步提高分數?

```python
# 顯示 1stFlrSF 與目標值的散佈圖
import seaborn as sns
import matplotlib.pyplot as plt

# 加這個的目的[:train_num]？？？
sns.regplot(x = df['1stFlrSF'][:train_num], y=train_Y) # reference:https://stackoverflow.com/questions/36317652/whats-the-meaning-of-x-y/36317764
plt.show

# 做線性迴歸, 觀察分數
train_X = MMEncoder.fit_transform(df)
estimator = LinearRegression()
cross_val_score(estimator, train_X, train_Y, cv=5).mean()
```

```python
# 將 1stFlrSF 限制在你覺得適合的範圍內, 調整離群值
"""
Your Code Here
"""

# 將 1stFlrSF 限制在 800 到 2500 以內, 調整離群值
df['1stFlrSF'] = df['1stFlrSF'].clip(800, 2200) # clip(800,2500)是<800的全部換成800，>2500的全部換成2500
sns.regplot(x = df['1stFlrSF'][:train_num], y=train_Y)
plt.show()


# 做線性迴歸, 觀察分數
train_X = MMEncoder.fit_transform(df)
estimator = LinearRegression()
cross_val_score(estimator, train_X, train_Y, cv=5).mean()
```

# 作業2
* 續前題, 去除離群值有兩類方式 :  捨棄離群值(刪除離群的資料) 以及調整離群值,  
請試著用同樣的上下限, 改為 '捨棄離群值' 的方法, 看看結果會變好還是變差? 並試著解釋原因。

- 自己Ans:結果變好，因為直接把部分資料捨棄，更專注在更集中的地方去算準確率就變好
- 專家Ans:可以再進一步提高分數，但捨棄離群資料時, 因為也同時減少了驗證的資料, 因此捨棄離群值未必比調整離群值好，實際結果仍需再驗證

```python
# 將 1stFlrSF 限制在你覺得適合的範圍內, 捨棄離群值
"""
Your Code Here
"""

# 將 1stFlrSF 限制在 800 到 2500 以內, 捨棄離群值
# (df['1stFlrSF']>800) & (df['1stFlrSF']< 250) 執行結果為True、False
keep_indexs = (df['1stFlrSF']> 800) & (df['1stFlrSF']< 2200) # <800的都捨棄，>2500得也都捨棄
# df[keep_indexs]
df = df[keep_indexs] # 執行結果把df中是True的row留著，False的row捨棄，所以新df的row數量會變少
train_Y = train_Y[keep_indexs] # 執行結果把train_Y中對應到keep_indexs是True的row的train_Y的值留著，False的row的train_Y的值捨棄
sns.regplot(x = df['1stFlrSF'], y=train_Y) # 因為上面整個df已經經過True的row留著，False的row捨棄的步驟，所以已產生新的df，所以df['1stFlrSF']也只剩下True的row
plt.show()



# 做線性迴歸, 觀察分數
train_X = MMEncoder.fit_transform(df)
estimator = LinearRegression()
cross_val_score(estimator, train_X, train_Y, cv=5).mean()
```
